```python
import json
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Helper functions

```python
def read_json(json_path):
    with open(json_path) as f:
        return json.load(f)
```

---
# Generated Dataset

```python
elapsed_times_gendata_e1 = read_json('out/generated_dataset/elapsed_times_e1.json')
elapsed_times_gendata_e2 = read_json('out/generated_dataset/elapsed_times_e2.json')
```

### Experiment 1.1

```python
n_sizes = []
elapsed_times = {'2':[], '5':[], '10':[]} 

for k,v in elapsed_times_gendata_e1.items():
    n_sizes.append(int(k.split('_')[0]))
    elapsed_times[k.split('_')[1]].append(v['total'])

n_sizes = sorted(list(set(n_sizes)))
```

```python
fig = plt.figure(figsize=(15, 5))
ax = fig.add_subplot(111)

for k,v in elapsed_times.items():
    ax.plot([1, 2, 3], v, label="k="+k, linestyle='--', marker='o', linewidth=0.8)
    
# Hide the right and top spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# Only show ticks on the left and bottom spines
ax.yaxis.set_ticks_position('left')
ax.xaxis.set_ticks_position('bottom')

ax.set_xlabel('Number of data points')
ax.set_ylabel('Elapsed time (s)')

labels = [item.get_text() for item in ax.get_xticklabels()]
for idx,v in enumerate(labels):
    labels[idx] = ''
    
labels[1] = '10k'
labels[5] = '100k'
labels[9] = '1M'
ax.set_xticklabels(labels)

ax.tick_params(axis='both', which='both',length=0)

ax.legend(loc=2)
ax.autoscale
plt.title('Total elapsed time for different \n data sizes and # clusters', size=20)

plt.savefig('figs/elapsed_gendata_e1.png', dpi=300)
plt.show()
```

```python
n_sizes = []
elapsed_times = {'2':[], '5':[], '10':[]} 

for k,v in elapsed_times_gendata_e1.items():
    n_sizes.append(int(k.split('_')[0]))
    elapsed_times[k.split('_')[1]].append(np.mean(v['iter']))

n_sizes = sorted(list(set(n_sizes)))
```

### Experiment 1.2

```python
n_sizes = []
elapsed_times = {} 

for k,v in elapsed_times_gendata_e2.items():
    elapsed_times[k] = v['total']
```

```python
fig = plt.figure(figsize=(15, 5))
ax = fig.add_subplot(111)

colors = elapsed_times
colors = ["blue"] + ["#FF0066"] * 5

ax1 = ax.plot(list(elapsed_times.keys()), list(elapsed_times.values()), color="black", linestyle='--', linewidth=0.8)
ax2 = ax.scatter(list(elapsed_times.keys()), list(elapsed_times.values()), color=colors, marker='o')


# Hide the right and top spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# Only show ticks on the left and bottom spines
ax.yaxis.set_ticks_position('left')
ax.xaxis.set_ticks_position('bottom')

ax.set_xlabel('Number of nodes')
ax.set_ylabel('Elapsed time (s)')
ax.tick_params(axis='both', which='both',length=0)


ax.text(list(elapsed_times.keys())[0], list(elapsed_times.values())[0] + 6.5, 'Serialized', color='black', 
        bbox=dict(facecolor='none', edgecolor='black', boxstyle='round,pad=0.4'))

plt.title('Total elapsed time vs. # nodes', size=20)

# plt.savefig('figs/elapsed_gendata_e2.png', dpi=300)
plt.show()
```
